1.  What is the difference between a function and a method in Python?

  Python, the terms function and method are related but have distinct differences based on their context and usage:

 ***Function***:

A function is a block of code that is designed to perform a specific task. It can be defined using the def keyword and can be called independently.

Functions can exist outside of classes (i.e., in the global scope or in a module).

A function is not tied to any object or class.

Example of a function:

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```

 ***Method***

A method is a function that is associated with an object (an instance of a class) or a class itself (in the case of class methods).

Methods are always called on an object (instance method) or a class (class method). The first parameter in an instance method is typically self, which refers to the instance of the class.

Example of a method:

```python
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"

p = Person("Alice")
print(p.greet())  # 'greet' is a method of the Person class
```

2.  Explain the concept of function arguments and parameters in Python.

 In Python, function arguments and parameters are closely related concepts that refer to the values used in a function call and the variable names used in a function definition, respectively. Here's an explanation of each:

 ***Function Parameters*** :

Parameters are the names defined in the function definition that accept values when the function is called.

They act as placeholders for the values that will be passed into the function.

Example:

```python
def greet(name, age):  # 'name' and 'age' are parameters
    return f"Hello {name}, you are {age} years old."
```

 ***Function Arguments*** :

Arguments are the actual values passed to the function when it is called.

These values are assigned to the corresponding parameters in the function definition.

Example:

```python
greet("Alice", 30)  # "Alice" and 30 are arguments
```

####Types of Arguments:

Python supports several types of arguments:

a. Positional Arguments:

These are the most common type of arguments.

The values are passed to the parameters in the order they appear in the function definition.

Example:

```python
def add(a, b):
    return a + b

print(add(3, 5))  # 3 and 5 are positional arguments
```

b. Keyword Arguments:

These are arguments passed by explicitly specifying the parameter names in the function call.

The order doesn't matter when using keyword arguments.

Example:

```python
def greet(name, age):
    return f"Hello {name}, you are {age} years old."

print(greet(age=30, name="Alice"))  # 'name' and 'age' are specified by keyword
```

c. Default Arguments:

You can define default values for parameters in the function. If no argument is provided for a parameter, the default value is used.

Example:

```python
def greet(name, age=25):
    return f"Hello {name}, you are {age} years old."

print(greet("Alice"))  # Uses default age of 25
print(greet("Bob", 30))  # Uses provided age of 30
```

d. Arbitrary Positional Arguments (*args):

*args allows a function to accept any number of positional arguments as a tuple.

This is useful when you're not sure how many arguments will be passed to the function.

Example:

```python
def sum_numbers(*args):
    return sum(args)

print(sum_numbers(1, 2, 3, 4))  # 10
print(sum_numbers(5, 6))  # 11
```

e. Arbitrary Keyword Arguments (**kwargs):

**kwargs allows a function to accept any number of keyword arguments as a dictionary.

This is useful when you want to handle named arguments dynamically.

Example:

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Alice", age=30)
# Output:
# name: Alice
# age: 30
```

3. What are the different ways to define and call a function in Python?

- Basic Function: Standard function definition and call using def.

- Default Arguments: Functions with default values for parameters.

- Arbitrary Positional Arguments (*args): Functions that accept an arbitrary number of positional arguments.

- Arbitrary Keyword Arguments (**kwargs): Functions that accept an arbitrary number of keyword arguments.

- Lambda Functions: Anonymous functions for simple one-liners.

- Passing Functions as Arguments: Functions can be passed as arguments to other functions.

- Multiple Return Values: Functions can return multiple values, often unpacked at the call site.

- Recursive Functions: Functions that call themselves, often used in problems like factorials, tree traversal, etc.

Let's define a function that calculates the area of different shapes. It should handle:

A circle (using its radius).

A rectangle (using its length and width).

A triangle (using its base and height). We will use default arguments and arbitrary keyword arguments (**kwargs) to make the function versatile.

Example:

```python
import math

def calculate_area(shape, **kwargs):
    if shape == "circle":
        radius = kwargs.get("radius", 1)  # Default radius is 1
        return math.pi * radius ** 2
    elif shape == "rectangle":
        length = kwargs.get("length", 1)  # Default length is 1
        width = kwargs.get("width", 1)    # Default width is 1
        return length * width
    elif shape == "triangle":
        base = kwargs.get("base", 1)      # Default base is 1
        height = kwargs.get("height", 1)  # Default height is 1
        return 0.5 * base * height
    else:
        return "Shape not recognized"

# Calling the function
print(calculate_area("circle", radius=5))  # Area of circle with radius 5
print(calculate_area("rectangle", length=4, width=6))  # Area of rectangle with length 4 and width 6
print(calculate_area("triangle", base=3, height=4))  # Area of triangle with base 3 and height 4
print(calculate_area("circle"))  # Default circle with radius 1
```

4. What is the purpose of the `return` statement in a Python function?

 The return statement in a Python function serves the purpose of sending a value back to the caller and terminating the function's execution. It allows a function to produce a result that can be used elsewhere in the program. Without a return statement, a function would not produce a result explicitly and would return None by default.

Key Purposes:
End the Function: When the return statement is encountered, the function stops executing, and control is returned to the place where the function was called.

Provide a Result: The value or object passed to the return statement is sent back to the caller, where it can be assigned to a variable or used in further computations.

Optional: If no return statement is used, Python functions return None by default.

Example:

```python
def square(number):
    result = number * number
    return result  # Returning the squared value

# Calling the function and storing the result
squared_value = square(4)
print(squared_value)  # Output: 16
```

5. What are iterators in Python and how do they differ from iterables?

In Python, iterators and iterables are concepts related to looping and traversing through collections (such as lists, tuples, and other sequences). They are both used in for-loops and other contexts where repeated access to elements is needed. However, they have distinct characteristics and behaviors.

Iterable:
An iterable is any Python object that can be looped over or iterated over, meaning it can return one item at a time when asked. An iterable is any object that implements the __iter__() method or has a __getitem__() method that supports sequential access to its elements.

Examples of Iterables: Lists, tuples, strings, dictionaries, sets, etc.

What makes something iterable: An object is considered iterable if it can produce an iterator.

Iterator:
An iterator is an object that actually performs the iteration. It is an object that represents a stream of data, and it keeps track of the current position during the iteration. An iterator implements two key methods:

__iter__(): Returns the iterator object itself.

__next__(): Returns the next value in the sequence. If there are no more items to return, it raises the StopIteration exception to signal the end of the iteration.

So, while an iterable can be iterated over, an iterator is the object that does the actual iteration.

Key Differences:
Iterable: An object that can return an iterator (e.g., lists, strings).

Iterator: An object that keeps track of the iteration state and can return the next item.

Example:


```python
# Iterable (a list)
numbers = [1, 2, 3]

# Create an iterator from the iterable
iterator = iter(numbers)

# Use the iterator to loop through the items
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
print(next(iterator))  # Output: 3

# If we try to call next() again, it will raise StopIteration
# print(next(iterator))  # This would raise StopIteration
```

6.  Explain the concept of generators in Python and how they are defined.

***Concept of Generators in Python***:

A generator in Python is a special type of iterator that is defined using a function but produces values one at a time, lazily. Unlike regular functions that return a value and exit, a generator function can pause execution and yield values one by one, maintaining its state between each call.

Generators allow for lazy evaluation, which means they generate items on the fly, as you iterate over them, rather than computing and storing all items in memory at once. This can be very memory efficient when working with large datasets or infinite sequences.

***Key Points about Generators*** :

- Generator Function: A function that contains one or more yield statements. It returns an iterator (a generator) instead of a single value.

- yield Statement: The yield keyword is used to produce a value from the generator function. When yield is called, the state of the function is "paused," and the value is sent back to the caller. The function can then resume from where it left off when the next value is requested.

- Lazy Evaluation: Generators compute values on demand and do not store the entire sequence in memory. This makes them efficient for handling large data sets or infinite sequences.

***Difference between Regular Functions and Generators*** :

Regular Functions: Use return to send a value back to the caller and terminate execution.

Generator Functions: Use yield to send a value, but execution can be paused, allowing the function to continue producing values when requested.

***How Generators Are Defined*** :

Generators are defined in a similar way to regular functions but use the yield statement instead of return.

Example:

```python
def count_up_to(limit):
    count = 1
    while count <= limit:
        yield count  # Yield the current count value
        count += 1

# Create a generator by calling the generator function
counter = count_up_to(5)

# Iterate through the generator using a for-loop
for num in counter:
    print(num)
```

7.  What are the advantages of using generators over regular functions?

 Generators offer several advantages over regular functions in Python, especially in terms of memory efficiency, performance, and flexibility. Here's a detailed breakdown of these advantages:

-  Memory Efficiency:

Generators compute values one at a time, lazily, meaning they don't store the entire sequence in memory. This makes them extremely memory-efficient, especially when dealing with large datasets or infinite sequences.

Regular functions that return lists or other collections typically store the entire result in memory, which can be inefficient for large datasets.

Example:
If you need to generate a sequence of numbers from 1 to 1 million:

A regular function would return a full list, requiring memory to store all 1 million numbers.

A generator would yield each number one at a time without holding the entire sequence in memory.

- Lazy Evaluation:

Generators generate values on demand, meaning they only compute and return a value when requested. This lazy evaluation ensures that the computation happens only when necessary.

Regular functions that return lists or other iterables compute and return all values upfront, which could be inefficient if you only need part of the sequence.

Example:
If you're processing a large file and only need to read certain lines, a generator can be used to read and yield lines one by one without loading the entire file into memory.

- Improved Performance:

Generators are generally faster than creating large collections (like lists) when the entire collection is not needed at once.

Since generators yield values one at a time and don’t need to build an entire structure (e.g., a list), they reduce the overhead of memory allocation and copying that is required when constructing large collections.

- Supports Infinite Sequences:

Generators can be used to represent infinite sequences, as they don't need to compute and store the entire sequence upfront.

Regular functions, on the other hand, cannot generate infinite sequences because they require the entire sequence to be stored in memory.

Example of an Infinite Generator:

```python
def infinite_counter():
    count = 1
    while True:
        yield count
        count += 1
```

- Simpler Code:

Generators often allow for cleaner and simpler code compared to the equivalent code that would use regular functions and return collections.

Generators handle the iteration logic internally with the yield statement, making the code more concise.

Example:
To generate a Fibonacci sequence:

With a regular function, you might need to manage the state explicitly (keeping track of the current and previous values).

With a generator, the state is maintained automatically, and the code remains simple.


```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
```

- Stateful Iteration:

Generators automatically maintain their state between successive calls. After each yield, the function "remembers" where it left off and resumes from that point the next time it is called.

In contrast, regular functions must rely on external mechanisms (like a separate variable) to maintain state across calls.

-  Cleaner for Complex Iteration:

Generators are ideal for scenarios where complex iteration logic is needed. Using yield, you can implement iterators without having to manually handle the iteration logic (e.g., incrementing indices, checking conditions).

- Better Resource Management:

Generators can be used to manage resources like file handles or network connections efficiently. They can yield data one piece at a time and release resources when they are no longer needed.

Regular functions that return lists or other data structures typically hold onto resources (like open files) until the entire result is produced.



8. What is a lambda function in Python and when is it typically used?

 Lambda Function in Python:

A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike a regular function, which is defined using the def keyword, a lambda function is a one-liner function that can have any number of arguments but only one expression. The expression is evaluated and returned automatically.

Syntax of Lambda Function:

```python
lambda arguments: expression
```

- lambda: Keyword to define the lambda function.

- arguments: A list of parameters passed to the lambda function.

- expression: A single expression that gets evaluated and returned. The expression cannot contain multiple statements.

Key Characteristics of Lambda Functions:

Anonymous: Lambda functions are often used for short, simple tasks and don't necessarily need a name.

Single Expression: A lambda function can only contain one expression. It cannot contain statements or multiple expressions.

Inline Usage: Lambda functions are typically used in places where you need a function for a short period of time, such as in functional programming constructs like map(), filter(), and reduce().

When are Lambda Functions Typically Used?:
Short Functions: When you need a quick, one-line function.

As Arguments to Higher-Order Functions: They are frequently used as arguments to functions like map(), filter(), sorted(), and others, which require a function as input.

Anonymous Functions: When you don’t need to define a full function and just want a simple function for a short task.

Example of a Lambda Function:
Let’s look at a simple example to understand how lambda functions work:

```python
# Regular function to add two numbers
def add(x, y):
    return x + y

# Lambda function to add two numbers
add_lambda = lambda x, y: x + y

# Using both functions
print(add(3, 5))          # Output: 8
print(add_lambda(3, 5))   # Output: 8
```

9. Explain the purpose and usage of the `map()` function in Python.

  The map() function helps you apply a function to each item in a list (or any other iterable, like a tuple). It returns an iterator (a collection that you can loop through), where each item is the result of applying the function to the items in the original list.

***How does it work?***

You give it a function and an iterable (like a list).

It applies that function to every item in the iterable.

It returns a new iterator with the updated values.

Simple Example:

Let’s say you have a list of numbers and you want to double each number. Here's how you can use map():


```python
# List of numbers
numbers = [1, 2, 3, 4, 5]

# Function to double a number
def double(num):
    return num * 2

# Use map() to apply the function to each number
doubled_numbers = map(double, numbers)

# Convert the map object to a list to see the result
print(list(doubled_numbers))  # Output: [2, 4, 6, 8, 10]
```

10.  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

 In Python, map(), reduce(), and filter() are all higher-order functions that work on iterables (like lists, tuples, etc.). They allow you to perform operations on elements of an iterable in a functional programming style. Let’s break them down:

-  map() Function:

Purpose: The map() function applies a given function to every item in an iterable (like a list or tuple) and returns a new iterable (an iterator) with the transformed values.

Usage: It's used when you want to modify or transform all items in an iterable in some way (e.g., applying a mathematical operation to each number).

Example of map():
Let’s say you want to square each number in a list:


```python
# List of numbers
numbers = [1, 2, 3, 4]

# Use map() to square each number
squared_numbers = map(lambda x: x ** 2, numbers)

# Convert the map object to a list to see the result
print(list(squared_numbers))  # Output: [1, 4, 9, 16]
```

- reduce() Function:

Purpose: The reduce() function takes a function and an iterable, and reduces the iterable to a single value by applying the function cumulatively to the elements.

Usage: It's used when you want to combine the elements of an iterable into one single result, like summing all numbers or multiplying them together.

Note: You need to import reduce() from the functools module because it's not available by default.

Example of reduce():
Let’s say you want to find the product of all numbers in a list:

```python
from functools import reduce

# List of numbers
numbers = [1, 2, 3, 4]

# Use reduce() to multiply all numbers together
product = reduce(lambda x, y: x * y, numbers)

print(product)  # Output: 24
```

- filter() Function:

Purpose: The filter() function filters elements from an iterable based on a condition (a function that returns True or False). It returns a new iterable with only the elements that satisfy the condition.

Usage: It's used when you want to filter out some items from an iterable based on a condition (e.g., selecting all even numbers).

Example of filter():
Let’s say you want to select all even numbers from a list:

```python
# List of numbers
numbers = [1, 2, 3, 4, 5, 6]

# Use filter() to get only the even numbers
even_numbers = filter(lambda x: x % 2 == 0, numbers)

# Convert the filter object to a list to see the result
print(list(even_numbers))  # Output: [2, 4, 6]
```

```python
def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime.")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.



```python
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            fib_number = self.a
            self.a, self.b = self.b, self.a + self.b  # Update for next Fibonacci number
            self.count += 1
            return fib_number
        else:
            raise StopIteration

# Example usage:
fib_iterator = FibonacciIterator(10)  # Generate first 10 Fibonacci numbers
for number in fib_iterator:
    print(number)
```

7.  Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line  # Yield each line as it is
    except FileNotFoundError:
        print(f"The file at {file_path} was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage:
file_path = 'example.txt'  # Replace with the path to your file
for line in read_file_line_by_line(file_path):
    print(line, end='')  # 'end' is used to avoid adding extra newlines
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius_temperatures = [0, 25, 30, 35, 40]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Use map() to apply the conversion function to each temperature in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the converted temperatures
print(fahrenheit_temperatures)
```
